```python
import matplotlib.pyplot as plt
import os
import numpy as np



# 定义一个时间窗口，这里假设时间窗口为5天
window_size = 50

# 创建文件夹保存生成的图像
output_folder = 'correlation_plots'
os.makedirs(output_folder, exist_ok=True)

target_list = ['Signal3','Signal5','Signal10','Signal20', 'Signal30']
summary_df = pd.DataFrame(columns=['col', 'Mean', 'Variance','min', 'max','target'])
for target in target_list:
# 创建空列表保存每个属性的相关性系数
    correlation_values_list = []
    top_correlations = pd.DataFrame()
    # 循环遍历数据框的每个时间点
    date_list = []
    for i in range(window_size, len(df), 50):
        # 获取当前时间点的数据
        subset_df = df.iloc[i - window_size:i + 1]

        # 计算相关性矩阵df.drop(['tradestatus', 'isST'], axis=1)
        correlation_matrix = subset_df.drop(['tradestatus', 'isST'], axis=1).corr()

        # 获取与'signal5'相关性系数的绝对值，并按降序排序
        correlations_with_signal5 = correlation_matrix[target].abs().drop(['Signal3', 'Signal5', 'Signal10', 'Signal20', 'Signal30'])

        # 将top 10相关性系数的列添加到top_correlations中
        top_10_columns = correlations_with_signal5.index
        top_correlations[f'Time {i}'] = top_10_columns
  
        # 获取具体的相关性系数
        correlation_values = {col: correlation_matrix.loc[target, col] for col in top_10_columns}
        
        # 将相关性系数字典添加到总列表中
        date_list.append(df.loc[i + 1]['date'])
        correlation_values_list.append(correlation_values)


    # 遍历每个相关性系数列
    for col in top_10_columns:

        correlation_values_for_col = []

        # Iterate over the correlation values list and collect values for the current column
        for correlation_values in correlation_values_list:
            if not np.isnan(correlation_values[col]):
                correlation_values_for_col.append(correlation_values[col] * 100)
        # Calculate and print mean and variance for the current column
        mean_value = np.mean(correlation_values_for_col)
        variance_value = np.var(correlation_values_for_col)
        max_value = np.max(np.abs(correlation_values_for_col))
        min_value = np.min(np.abs(correlation_values_for_col))
        new_row = {'col': col, 'Mean': mean_value, 'Variance':variance_value, 'min':min_value, 'max':max_value, 'target':target}
        summary_df = summary_df.append(new_row, ignore_index=True)
        print(f"{target} Column: {col}, Mean: {mean_value}, Variance: {variance_value}, max_value: {max_value}, min_value: {min_value}")
        # 绘制折线图
        plt.figure(figsize=(20, 8))
        for i, correlation_values in enumerate(correlation_values_list):
            plt.plot(date_list[i], correlation_values[col], 'o-',linewidth=1)
        plt.xlabel('Time')
        plt.ylabel('Correlation Value')
        plt.title(f'Correlation Values Over Time - {col}')
        plt.axhline(0, color='gray', linewidth=0.8)
        plt.xticks(rotation=45)
        plt.tight_layout()

        # 保存折线图
        output_file = os.path.join(output_folder, f'{target}_{col}_correlation_plot.png')
        plt.savefig(output_file)
        plt.close()
```
